# Imputación de datos

```python
df3.isnull()   #"isnull" ayuda a conocer los datos con caracteríscitas "NaN o Null"
```

```python
df3.isnull().sum()/len(df3)   #Realiza un conteo de las variables "NaN o Null" del data frame. Luego se divide por la longitud de su varible
```

```python
df3.info()   #Tipo de variables de los datos per columna
```

```python
df3['saldo'].fillna(0, inplace=True)   #"fillna" reemplaza los valores "NaN or NULL". En este caso los reemplazará por cero (0)
df3['saldo'].isnull().sum()     #Realiza nuevamente el conteo de las variables "NaN o Null", especificamente para la columna "saldo"
```

```python
df3['saldo'].min()    #Conocer el valor mínimo de la columna
```

Retirar valores negativos de la variable "antigemp"

```python
df3['antigemp'].value_counts()   #Contar cantidad de valores
```

```python
df3['antigemp']<=0   #Conocer los valores menores e iguales e iguales a cero (0)
```

```python
#".index" almacena como etiqueta los valores. En este caso, llama los valores iguales a 0
index1 = df3[df3['antigemp']<=0].index
index1
```

```python
df3.loc[index1,'antigemp']=1   #".loc"Accede al grupo de filas o columnas etiquetadas como "true" y las modifica por el valor de 1
```

```python
mask = df3['antigemp']>=0
mask
```

```python
# Eliminar los valores menores a cero
mask = df3['antigemp']>=0
df3=df3[mask]
df3['antigemp'].value_counts()
```

```python
df3['antigemp'].fillna(df3.antigemp.mean(), inplace=True)   #Remplazar valores "NaN o Null" por la media
df3['antigemp'].isnull().sum()
```
